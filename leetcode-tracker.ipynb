```python
import pandas as pd
```

```python
df = pd.read_csv('questions.csv')
df_dict = df.to_dict('records')

for item in df_dict:
    item['Topic_Tags'] = eval(item['Topic_Tags'])

df = pd.DataFrame(df_dict)
```

```python
df
```

```python
all_topic_tags = list(set(tag for tags in df['Topic_Tags'] for tag in tags))
all_topic_tags
```

```python
topic_tag_dict = {}

# Loop through each row and each Topic_Tag
for index, row in df.iterrows():
    for tag in row['Topic_Tags']:
        if tag not in topic_tag_dict:
            topic_tag_dict[tag] = []
        topic_tag_dict[tag].append(row['Question_No'])

# Display the dictionary
print(topic_tag_dict)
```

```python
topic_tag_df = pd.DataFrame.from_dict(topic_tag_dict, orient='index')
topic_tag_df
```
